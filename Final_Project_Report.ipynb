The overall analytical framework for this project follows a structured and iterative machine learning pipeline designed to evaluate model performance, understand the impact of feature transformations, and identify the best-performing model for predicting housing prices.

The workflow consists of four main stages:

**Baseline Model Evaluation**
We first established baseline performance using three models: Ridge Regression, Random Forest, and Gradient Boosting. This step provided a reference point for evaluating improvements introduced in later stages and allowed us to assess whether the problem exhibits linear or nonlinear behavior.

**Feature Engineering and Feature Selection**
We then explored feature engineering to determine whether additional transformations or interactions could improve predictive performance. This was followed by feature selection to evaluate whether reducing the number of predictors could enhance model efficiency and generalization.
**Model Optimization through Hyperparameter Tuning*
Hyperparameter tuning was performed using GridSearchCV and RandomizedSearchCV to identify optimal parameter configurations. This step was critical for improving the performance of tree-based models, particularly Gradient Boosting, which is sensitive to parameters such as learning rate and tree depth.

**Final Model Evaluation**
The best-performing model was selected based on cross-validation performance and evaluated on a held-out test set to assess generalization.

**Use of Validation Curves**

Validation curves were used implicitly through cross-validation during model evaluation and hyperparameter tuning. By evaluating model performance across different folds and parameter configurations, we ensured that the selected models generalized well to unseen data and avoided overfitting.

In particular, hyperparameter tuning allowed us to analyze how model performance varied with parameters such as:

Tree depth
Number of estimators
Learning rate

This provided insight into the trade-off between model complexity and generalization.

**Choice of Error Metric**

The primary error metric used in this project is Mean Absolute Error (MAE).

MAE was chosen because:

It provides a direct and interpretable measure of prediction error in the same units as the target variable (price)
It is less sensitive to outliers than RMSE, which is important given the presence of extreme housing prices in the dataset
It aligns well with real-world objectives, where absolute prediction error is often more meaningful than squared error

Given the skewness and variability of housing prices, MAE provides a more robust and reliable measure of model performance compared to RMSE.

### 4.3 Feature Engineering

Describe your feature engineering.

Include:
- transformations or new features
- why you created them
- which were useful or not useful
- what worked and what did not work

**4.3 Answer:**  

Feature engineering in this project was guided by domain knowledge of real estate rather than exhaustive trial-and-error. The objective was to construct transformations that are interpretable, meaningful, and capable of capturing nonlinear relationships in housing data.

A logarithmic transformation of living area (`sqft_log`) was introduced to account for the diminishing marginal impact of size on property value. In real estate markets, increases in square footage do not translate linearly into price increases, and the log transformation helps stabilize this relationship while reducing skewness.

The bathroom-to-bedroom ratio (`bath_bed_ratio`) was included to capture the functional quality and usability of a property. Homes with a higher number of bathrooms relative to bedrooms tend to be more desirable, making this a more informative measure than simple counts.

Home age (`home_age`) was derived from the year built to reflect depreciation and modernization effects. Newer homes typically command higher prices due to updated features, while older homes may decrease in value unless renovated.

The living area to lot size ratio (`living_to_lot_ratio`) was created to represent land utilization. This feature distinguishes between properties with similar living areas but different lot sizes, capturing whether a property prioritizes interior space or land.

These engineered features provide complementary perspectives to the original variables and align with established practices in real estate modeling. In particular, they enhance interpretability and offer more intuitive representations of property characteristics.

From a modeling perspective, the results indicate that while these transformations are meaningful, their impact on predictive performance is modest when using tree-based ensemble models such as Random Forest and Gradient Boosting. This outcome suggests that these models are already effective at capturing nonlinear relationships and feature interactions directly from the original variables.

Overall, feature engineering in this project serves to reinforce domain understanding and improve interpretability, while also demonstrating that advanced ensemble methods can internally learn complex relationships without requiring extensive manual transformations.

### 4.4 Model Selection

The model selection process in this project followed a structured and data-driven approach, evaluating multiple models and refining them based on performance and generalization.

  A sequential ensemble model that iteratively improves predictions by correcting previous errors, allowing for high predictive accuracy.

This combination enabled a clear comparison between linear and nonlinear modeling approaches.

#### Evaluation of Generalization

Model performance was evaluated using **cross-validation with Mean Absolute Error (MAE)** as the primary metric.

Cross-validation was used to:
- Provide reliable performance estimates across multiple data splits  
- Reduce variance in model evaluation  
- Assess how well models generalize to unseen data  

Additionally, the final model was evaluated on a **held-out test set**.

- **Final CV MAE:** 135,701.835 ± 1,337.701  
- **Test MAE:** 136,665.406  

The close alignment between these values indicates strong generalization and minimal overfitting.
